# Image1

```python
import matplotlib.pyplot as plt # shorthand for pyplot in matplotlib, plotting
import numpy as np # another shorthand, arrays
from PIL import Image # read in images
```

```python
im1 = Image.open("image1.jpg")
```

```python
im1_data = np.array(im1)
```

```python
im1_data.shape
```

```python
plt.imshow(im1_data) # quick look with matplotlib
```

```python
fig, ax = plt.subplots(figsize=(10,3))
ax.hist(im1_data[:,:,0]) # [ : means all of the rows, : all of the columns, 0th/1st/Red channel]
plt.show()
```

```python
fig, axes = plt.subplots(4,1,figsize=(10,9)) # 4 rows, instead of 3
axes[0].hist(im1_data[:,:,0].flatten(),bins=256, color='red') # 0th, R channel
axes[1].hist(im1_data[:,:,1].flatten(),bins=256, color='green') # 1st, G channel
axes[2].hist(im1_data[:,:,2].flatten(),bins=256, color='blue') # 2nd, B channel

axes[0].set_xlabel('Red Bins')
axes[1].set_xlabel('Green Bins')
axes[2].set_xlabel('Blue Bins')

axes[0].set_ylabel('# pixels')
axes[1].set_ylabel('# pixels')
axes[2].set_ylabel('# pixels')

# image in last row
axes[3].imshow(im1_data)

plt.show()
```

## Discussion of Image1 and its color distribution.

According to the histogram of image 1, the three colors seem to have similar amounts and concentrate around 200. I think this because the ground is nearly white. The blue has significantly more pixels of 0 than the red and the green because the dog is nearly yellow. However, it is strange that the green has fewer pixels of 255 than the red. There are meadows in the background, and this should cause more green in 255 than red.
